## Problema 21: Prueba de Hipótesis para el Peso Promedio de Paquetes de Cereal

**Enunciado:** Una empresa afirma que sus paquetes de cereal tienen un peso promedio de 500 g. Se sospecha que la máquina está descalibrada y subllenando. Se toma una muestra de 40 paquetes, obteniendo una media de 495 g. La desviación estándar poblacional es σ = 15 g. Con un nivel de de significancia α = 0,05.
1. Plantee las hipótesis nula y alternativa.
2. Calcule el estadístico Z.
3. Determine la región crítica y la región de rechazo.
4. Concluya si se rechaza H0.

### 1. Plantear las hipótesis nula y alternativa

La afirmación de la empresa es que el peso promedio es de 500 g. La sospecha es que la máquina está subllenando, lo que implica que el peso promedio real es menor a 500 g.

*   **Hipótesis Nula (H₀):** El peso promedio de los paquetes de cereal es igual a 500 g. (µ = 500)
*   **Hipótesis Alternativa (H₁):** El peso promedio de los paquetes de cereal es menor a 500 g. (µ < 500)

Este es un problema de prueba de hipótesis unilateral (cola izquierda).

### 2. Establecer el nivel de significancia α

El nivel de significancia dado es α = 0.05.

### 3. Identificar el estadístico de prueba y su distribución

Como la desviación estándar poblacional (σ) es conocida y el tamaño de la muestra es grande (n = 40 > 30), utilizaremos el **estadístico Z** y la **distribución normal estándar**.

### 4. Determinar el valor crítico y la región de rechazo

Para una prueba unilateral de cola izquierda con α = 0.05, necesitamos encontrar el valor Z crítico tal que el área a su izquierda sea 0.05.

La región de rechazo estará en la cola izquierda de la distribución normal estándar. Rechazaremos H₀ si el estadístico Z calculado es menor que el valor Z crítico.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
sample_mean = 495  # Media muestral
population_mean_h0 = 500 # Media poblacional bajo H0
population_std_dev = 15 # Desviación estándar poblacional (sigma)
sample_size = 40   # Tamaño de la muestra
alpha = 0.05       # Nivel de significancia

# Calcular el error estándar
# Fórmula: SE = sigma / sqrt(n)
standard_error = population_std_dev / np.sqrt(sample_size)
print(f"Error estándar (SE): {standard_error:.4f}")

# Encontrar el valor crítico Z para una prueba unilateral de cola izquierda con alpha = 0.05
# Usamos stats.norm.ppf() para obtener el valor Z cuya área acumulada a la izquierda es alpha.
critical_z = stats.norm.ppf(alpha)
print(f"Valor crítico Z para alpha = {alpha}: {critical_z:.4f}")

# Calcular el estadístico de prueba Z
# Fórmula: Z = (media_muestral - media_poblacional_H0) / SE
z_statistic = (sample_mean - population_mean_h0) / standard_error
print(f"Estadístico Z calculado: {z_statistic:.4f}")

# Calcular el valor p (para una prueba unilateral de cola izquierda)
# El valor p es el área a la izquierda del estadístico Z calculado.
p_value = stats.norm.cdf(z_statistic)
print(f"Valor p: {p_value:.4f}")

# Graficar la distribución normal estándar
x = np.linspace(-4, 4, 200)
y = stats.norm.pdf(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Distribución Normal Estándar con Región de Rechazo")
ax.set_xlabel("Valor Z")
ax.set_ylabel("Densidad de Probabilidad")
ax.grid(True)

# Rellenar la región de rechazo (cola izquierda)
x_reject = np.linspace(-4, critical_z, 100)
y_reject = stats.norm.pdf(x_reject)
ax.fill_between(x_reject, y_reject, color='purple', alpha=0.5, label=f'Región de Rechazo (α = {alpha})')

# Marcar el valor crítico Z
ax.axvline(critical_z, color='purple', linestyle='--', label=f'Valor Crítico Z = {critical_z:.4f}')

# Marcar el estadístico Z calculado
ax.axvline(z_statistic, color='black', linestyle='-', label=f'Estadístico Z Calculado = {z_statistic:.4f}')

# Agregar leyenda
ax.legend()

plt.show()

# Tomar una decisión estadística
print("\n--- Decisión Estadística ---")
if z_statistic < critical_z:
    print(f"El estadístico Z calculado ({z_statistic:.4f}) es menor que el valor crítico Z ({critical_z:.4f}).")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El estadístico Z calculado ({z_statistic:.4f}) NO es menor que el valor crítico Z ({critical_z:.4f}).")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")

# Tomar una decisión basada en el valor p
print("\n--- Decisión basada en el Valor p ---")
if p_value < alpha:
    print(f"El valor p ({p_value:.4f}) es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El valor p ({p_value:.4f}) NO es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")


# Conclusión en términos prácticos
print("\n--- Conclusión ---")
if z_statistic < critical_z:
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que el peso promedio de los paquetes de cereal es significativamente menor a 500 g. Esto sugiere que la máquina podría estar descalibrada y subllenando los paquetes.")
    print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una media muestral tan baja o más baja como 495 g, si la verdadera media poblacional fuera 500 g. Un valor p pequeño ({p_value:.4f} < {alpha}) sugiere que este resultado es improbable si la hipótesis nula fuera cierta.")
else:
     print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que el peso promedio de los paquetes de cereal es significativamente menor a 500 g. No podemos rechazar la afirmación de que el peso promedio es de 500 g. Esto no significa que la media sea exactamente 500 g, sino que la muestra no proporciona evidencia fuerte en contra de H0.")
     print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una media muestral tan baja o más baja como 495 g, si la verdadera media poblacional fuera 500 g. Un valor p grande ({p_value:.4f} >= {alpha}) sugiere que este resultado es plausible si la hipótesis nula fuera cierta.")
```

Para una prueba unilateral de cola izquierda con α = 0.01, necesitamos encontrar el valor Z crítico tal que el área a su izquierda sea 0.01.

La región de rechazo estará en la cola izquierda de la distribución normal estándar. Rechazaremos H₀ si el estadístico Z calculado es menor que el valor Z crítico.

## Problema 23: Prueba de Hipótesis para el Promedio de Clase en un Examen Nacional

**Enunciado:** Un profesor afirma que el promedio de su clase en un examen nacional es diferente al promedio nacional de 75 puntos. Una muestra de 36 estudiantes de su clase tiene una media de 78. La desviación estándar poblacional es σ = 12. Con un nivel de significancia: α = 0,05.
1. Plantee las hipótesis nula y alternativa.
2. Calcule el estadístico Z.
3. Determine la región crítica y la región de rechazo.
4. Concluya si se rechaza H0.

### 1. Plantear las hipótesis nula y alternativa

La afirmación del profesor es que el promedio de su clase es diferente al promedio nacional de 75 puntos.

*   **Hipótesis Nula (H₀):** El promedio de la clase es igual al promedio nacional de 75 puntos. (µ = 75)
*   **Hipótesis Alternativa (H₁):** El promedio de la clase es diferente al promedio nacional de 75 puntos. (µ ≠ 75)

Este es un problema de prueba de hipótesis bilateral (dos colas).

### 2. Establecer el nivel de significancia α

El nivel de significancia dado es α = 0.05.

Para una prueba bilateral, el nivel de significancia se divide entre las dos colas: α/2 = 0.05 / 2 = 0.025.

### 3. Identificar el estadístico de prueba y su distribución

Como la desviación estándar poblacional (σ) es conocida y el tamaño de la muestra es grande (n = 36 > 30), utilizaremos el **estadístico Z** y la **distribución normal estándar**.

### 4. Determinar el valor crítico y la región de rechazo

Para una prueba bilateral con α = 0.05, necesitamos encontrar dos valores Z críticos: uno en la cola izquierda y uno en la cola derecha.

*   El valor Z crítico de la cola izquierda es tal que el área a su izquierda es α/2 = 0.025.
*   El valor Z crítico de la cola derecha es tal que el área a su derecha es α/2 = 0.025 (o el área a su izquierda es 1 - α/2 = 0.975).

La región de rechazo estará en ambas colas de la distribución normal estándar. Rechazaremos H₀ si el estadístico Z calculado es menor que el valor Z crítico de la cola izquierda O si el estadístico Z calculado es mayor que el valor Z crítico de la cola derecha.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
sample_mean = 78  # Media muestral
population_mean_h0 = 75 # Media poblacional bajo H0
population_std_dev = 12 # Desviación estándar poblacional (sigma)
sample_size = 36   # Tamaño de la muestra
alpha = 0.05       # Nivel de significancia
alpha_per_tail = alpha / 2 # Nivel de significancia por cola para prueba bilateral

# Calcular el error estándar
# Fórmula: SE = sigma / sqrt(n)
standard_error = population_std_dev / np.sqrt(sample_size)
print(f"Error estándar (SE): {standard_error:.4f}")

# Encontrar los valores críticos Z para una prueba bilateral con alpha = 0.05
# Valor crítico Z de la cola izquierda (área a la izquierda = alpha/2)
critical_z_lower = stats.norm.ppf(alpha_per_tail)
# Valor crítico Z de la cola derecha (área a la izquierda = 1 - alpha/2)
critical_z_upper = stats.norm.ppf(1 - alpha_per_tail)

print(f"Valores críticos Z para alpha = {alpha} (prueba bilateral): [{critical_z_lower:.4f}, {critical_z_upper:.4f}]")

# Calcular el estadístico de prueba Z
# Fórmula: Z = (media_muestral - media_poblacional_H0) / SE
z_statistic = (sample_mean - population_mean_h0) / standard_error
print(f"Estadístico Z calculado: {z_statistic:.4f}")

# Calcular el valor p para una prueba bilateral
# Para una prueba bilateral, el valor p es el doble del área en la cola
# correspondiente al estadístico de prueba.
# Si Z > 0, el valor p = 2 * (1 - stats.norm.cdf(Z))
# If Z < 0, el valor p = 2 * stats.norm.cdf(Z)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
print(f"Valor p (prueba bilateral): {p_value:.4f}")


# Graficar la distribución normal estándar
x = np.linspace(-4, 4, 200)
y = stats.norm.pdf(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Distribución Normal Estándar con Regiones de Rechazo (Bilateral)")
ax.set_xlabel("Valor Z")
ax.set_ylabel("Densidad de Probabilidad")
ax.grid(True)

# Rellenar las regiones de rechazo (dos colas)
x_reject_lower = np.linspace(-4, critical_z_lower, 100)
y_reject_lower = stats.norm.pdf(x_reject_lower)
ax.fill_between(x_reject_lower, y_reject_lower, color='purple', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_per_tail})')

x_reject_upper = np.linspace(critical_z_upper, 4, 100)
y_reject_upper = stats.norm.pdf(x_reject_upper)
ax.fill_between(x_reject_upper, y_reject_upper, color='purple', alpha=0.5) # No label for the second fill to avoid duplicate legend entries


# Marcar los valores críticos Z
ax.axvline(critical_z_lower, color='purple', linestyle='--', label=f'Valores Críticos Z = {critical_z_lower:.4f}, {critical_z_upper:.4f}')
ax.axvline(critical_z_upper, color='purple', linestyle='--')

# Marcar el estadístico Z calculado
ax.axvline(z_statistic, color='black', linestyle='-', label=f'Estadístico Z Calculado = {z_statistic:.4f}')

# Agregar leyenda
ax.legend()

plt.show()


# Tomar una decisión estadística
print("\n--- Decisión Estadística ---")
if z_statistic < critical_z_lower or z_statistic > critical_z_upper:
    print(f"El estadístico Z calculado ({z_statistic:.4f}) cae en la región de rechazo.")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El estadístico Z calculado ({z_statistic:.4f}) NO cae en la región de rechazo.")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")

# Tomar una decisión basada en el valor p
print("\n--- Decisión basada en el Valor p ---")
if p_value < alpha:
    print(f"El valor p ({p_value:.4f}) es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El valor p ({p_value:.4f}) NO es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")


# Conclusión en términos prácticos
print("\n--- Conclusión ---")
if p_value < alpha:
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que el promedio de la clase es significativamente diferente al promedio nacional de 75 puntos.")
    print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una media muestral tan extrema o más extrema como 78 g (en cualquier dirección), si el verdadero promedio nacional fuera 75 g. Un valor p pequeño ({p_value:.4f} < {alpha}) sugiere que este resultado es improbable si la hipótesis nula fuera cierta.")
else:
     print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que el promedio de la clase es significativamente diferente al promedio nacional de 75 puntos. No podemos rechazar la afirmación de que el promedio de la clase es igual al promedio nacional. Esto no significa que la media sea exactamente 75 puntos, sino que la muestra no proporciona evidencia fuerte en contra de H0.")
     print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una media muestral tan extrema o más extrema como 78 g (en cualquier dirección), si el verdadero promedio nacional fuera 75 g. Un valor p grande ({p_value:.4f} >= {alpha}) sugiere que este resultado es plausible si la hipótesis nula fuera cierta.")
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
sample_mean = 498  # Media muestral
population_mean_h0 = 500 # Media poblacional bajo H0
population_std_dev = 5 # Desviación estándar poblacional (sigma)
sample_size = 60   # Tamaño de la muestra
alpha = 0.01       # Nivel de significancia
alpha_per_tail = alpha / 2 # Nivel de significancia por cola para prueba bilateral

# Calcular el error estándar
# Fórmula: SE = sigma / sqrt(n)
standard_error = population_std_dev / np.sqrt(sample_size)
print(f"Error estándar (SE): {standard_error:.4f}")

# Encontrar los valores críticos Z para una prueba bilateral con alpha = 0.01
# Valor crítico Z de la cola izquierda (área a la izquierda = alpha/2)
critical_z_lower = stats.norm.ppf(alpha_per_tail)
# Valor crítico Z de la cola derecha (área a la izquierda = 1 - alpha/2)
critical_z_upper = stats.norm.ppf(1 - alpha_per_tail)

print(f"Valores críticos Z para alpha = {alpha} (prueba bilateral): [{critical_z_lower:.4f}, {critical_z_upper:.4f}]")

# Calcular el estadístico de prueba Z
# Fórmula: Z = (media_muestral - media_poblacional_H0) / SE
z_statistic = (sample_mean - population_mean_h0) / standard_error
print(f"Estadístico Z calculado: {z_statistic:.4f}")

# Calcular el valor p para una prueba bilateral
# Para una prueba bilateral, el valor p es el doble del área en la cola
# correspondiente al estadístico de prueba.
# Si Z > 0, el valor p = 2 * (1 - stats.norm.cdf(Z))
# If Z < 0, el valor p = 2 * stats.norm.cdf(Z)
p_value = 2 * stats.norm.cdf(z_statistic) # For Z_statistic < 0, p-value is 2 * area to the left
print(f"Valor p (prueba bilateral): {p_value:.4f}")


# Graficar la distribución normal estándar
x = np.linspace(-4, 4, 200)
y = stats.norm.pdf(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Distribución Normal Estándar con Regiones de Rechazo (Bilateral)")
ax.set_xlabel("Valor Z")
ax.set_ylabel("Densidad de Probabilidad")
ax.grid(True)

# Rellenar las regiones de rechazo (dos colas)
x_reject_lower = np.linspace(-4, critical_z_lower, 100)
y_reject_lower = stats.norm.pdf(x_reject_lower)
ax.fill_between(x_reject_lower, y_reject_lower, color='purple', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_per_tail})')

x_reject_upper = np.linspace(critical_z_upper, 4, 100)
y_reject_upper = stats.norm.pdf(x_reject_upper)
ax.fill_between(x_reject_upper, y_reject_upper, color='purple', alpha=0.5) # No label for the second fill to avoid duplicate legend entries


# Marcar los valores críticos Z
ax.axvline(critical_z_lower, color='purple', linestyle='--', label=f'Valores Críticos Z = {critical_z_lower:.4f}, {critical_z_upper:.4f}')
ax.axvline(critical_z_upper, color='purple', linestyle='--')

# Marcar el estadístico Z calculado
ax.axvline(z_statistic, color='black', linestyle='-', label=f'Estadístico Z Calculado = {z_statistic:.4f}')

# Agregar leyenda
ax.legend()

plt.show()


# Tomar una decisión estadística
print("\n--- Decisión Estadística ---")
if z_statistic < critical_z_lower or z_statistic > critical_z_upper:
    print(f"El estadístico Z calculado ({z_statistic:.4f}) cae en la región de rechazo.")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El estadístico Z calculado ({z_statistic:.4f}) NO cae en la región de rechazo.")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")

# Tomar una decisión basada en el valor p
print("\n--- Decisión basada en el Valor p ---")
if p_value < alpha:
    print(f"El valor p ({p_value:.4f}) es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El valor p ({p_value:.4f}) NO es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")


# Conclusión en términos prácticos
print("\n--- Conclusión ---")
if p_value < alpha:
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que el volumen promedio de las botellas llenadas por la máquina es significativamente diferente a 500 mL.")
    print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una media muestral tan extrema o más extrema como 498 mL (en cualquier dirección), si el verdadero volumen promedio fuera 500 mL. Un valor p pequeño ({p_value:.4f} < {alpha}) sugiere que este resultado es improbable si la hipótesis nula fuera cierta.")
else:
     print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que el volumen promedio de las botellas llenadas por la máquina es significativamente diferente a 500 mL. No podemos rechazar la afirmación de que el volumen promedio es de 500 mL. Esto no significa que la media sea exactamente 500 mL, sino que la muestra no proporciona evidencia fuerte en contra de H0.")
     print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una media muestral tan extrema o más extrema como 498 mL (en cualquier dirección), si el verdadero volumen promedio fuera 500 mL. Un valor p grande ({p_value:.4f} >= {alpha}) sugiere que este resultado es plausible si la hipótesis nula fuera cierta.")
```

## Problema 26: Prueba de Hipótesis para el Tiempo Promedio de Entrega a Domicilio

**Enunciado:** Un restaurante afirma que su tiempo promedio de entrega a domicilio es 25 minutos. Un cliente sospecha que es mayor. Se registran 15 entregas, obteniendo una media de 28 minutos y desviación estándar muestral s = 4 minutos. Con un nivel de significancia de α = 0,05.
1. Plantee las hipótesis nula y alternativa.
2. Calcule el estadístico t.
3. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de
rechazo.
4. Concluya si se rechaza H0.

### 1. Plantear las hipótesis nula y alternativa

La afirmación del restaurante es que el tiempo promedio de entrega es de 25 minutos. La sospecha del cliente es que el tiempo promedio real es mayor a 25 minutos.

*   **Hipótesis Nula (H₀):** El tiempo promedio de entrega es igual a 25 minutos. (µ = 25)
*   **Hipótesis Alternativa (H₁):** El tiempo promedio de entrega es mayor a 25 minutos. (µ > 25)

Este es un problema de prueba de hipótesis unilateral (cola derecha).

### 2. Establecer el nivel de significancia α

El nivel de significancia dado es α = 0.05.

### 3. Identificar el estadístico de prueba y su distribución

Como la desviación estándar poblacional (σ) es desconocida y el tamaño de la muestra es pequeño (n = 15 < 30), utilizaremos el **estadístico t** y la **distribución t de Student**.

### 4. Determinar los grados de libertad, el valor crítico t y la región de rechazo

Los grados de libertad para la distribución t se calculan como el tamaño de la muestra menos 1: df = n - 1.

Para una prueba unilateral de cola derecha con α = 0.05 y los grados de libertad correspondientes, necesitamos encontrar el valor t crítico tal que el área a su derecha sea 0.05.

La región de rechazo estará en la cola derecha de la distribución t. Rechazaremos H₀ si el estadístico t calculado es mayor que el valor t crítico.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
sample_mean = 28  # Media muestral
population_mean_h0 = 25 # Media poblacional bajo H0
sample_std_dev = 4 # Desviación estándar muestral (s)
sample_size = 15   # Tamaño de la muestra
alpha = 0.05       # Nivel de significancia

# Calcular los grados de libertad
degrees_of_freedom = sample_size - 1
print(f"Grados de libertad (df): {degrees_of_freedom}")

# Calcular el error estándar (usando desviación estándar muestral)
# Fórmula: SE = s / sqrt(n)
standard_error = sample_std_dev / np.sqrt(sample_size)
print(f"Error estándar (SE): {standard_error:.4f}")

# Encontrar el valor crítico t para una prueba unilateral de cola derecha con alpha = 0.05 y df
# Usamos stats.t.ppf() para obtener el valor t cuya área acumulada a la izquierda es 1 - alpha, con los grados de libertad apropiados.
critical_t = stats.t.ppf(1 - alpha, degrees_of_freedom)
print(f"Valor crítico t para alpha = {alpha} y df = {degrees_of_freedom}: {critical_t:.4f}")

# Calcular el estadístico de prueba t
# Fórmula: t = (media_muestral - media_poblacional_H0) / SE
t_statistic = (sample_mean - population_mean_h0) / standard_error
print(f"Estadístico t calculado: {t_statistic:.4f}")

# Calcular el valor p (para una prueba unilateral de cola derecha)
# El valor p es el área a la derecha del estadístico t calculado.
p_value = 1 - stats.t.cdf(t_statistic, degrees_of_freedom)
print(f"Valor p: {p_value:.4f}")

# Graficar la distribución t de Student
x = np.linspace(stats.t.ppf(0.001, degrees_of_freedom),
                stats.t.ppf(0.999, degrees_of_freedom), 200)
y = stats.t.pdf(x, degrees_of_freedom)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title(f"Distribución t de Student (df = {degrees_of_freedom}) con Región de Rechazo (Cola Derecha)")
ax.set_xlabel("Valor t")
ax.set_ylabel("Densidad de Probabilidad")
ax.grid(True)

# Rellenar la región de rechazo (cola derecha)
x_reject = np.linspace(critical_t, stats.t.ppf(0.999, degrees_of_freedom), 100)
y_reject = stats.t.pdf(x_reject, degrees_of_freedom)
ax.fill_between(x_reject, y_reject, color='red', alpha=0.5, label=f'Región de Rechazo (α = {alpha})')

# Marcar el valor crítico t
ax.axvline(critical_t, color='purple', linestyle='--', label=f'Valor Crítico t = {critical_t:.4f}')

# Marcar el estadístico t calculado
ax.axvline(t_statistic, color='black', linestyle='-', label=f'Estadístico t Calculado = {t_statistic:.4f}')

# Agregar leyenda
ax.legend()

plt.show()


# Tomar una decisión estadística
print("\n--- Decisión Estadística ---")
if t_statistic > critical_t:
    print(f"El estadístico t calculado ({t_statistic:.4f}) es mayor que el valor crítico t ({critical_t:.4f}).")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El estadístico t calculado ({t_statistic:.4f}) NO es mayor que el valor crítico t ({critical_t:.4f}).")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")

# Tomar una decisión basada en el valor p
print("\n--- Decisión basada en el Valor p ---")
if p_value < alpha:
    print(f"El valor p ({p_value:.4f}) es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El valor p ({p_value:.4f}) NO es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")

# Conclusión en términos prácticos
print("\n--- Conclusión ---")
if p_value < alpha:
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que el tiempo promedio de entrega a domicilio del restaurante es significativamente mayor a 25 minutos. Esto apoya la sospecha del cliente.")
    print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una media muestral tan alta o más alta como 28 minutos, si el verdadero tiempo promedio fuera 25 minutos. Un valor p pequeño ({p_value:.4f} < {alpha}) sugiere que este resultado es improbable si la hipótesis nula fuera cierta.")
else:
     print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que el tiempo promedio de entrega a domicilio del restaurante es significativamente mayor a 25 minutos. No podemos rechazar la afirmación de que el tiempo promedio es de 25 minutos. Esto no significa que la media sea exactamente 25 minutos, sino que la muestra no proporciona evidencia fuerte en contra de H0.")
     print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una media muestral tan alta o más alta como 28 minutos, si el verdadero tiempo promedio fuera 25 minutos. Un valor p grande ({p_value:.4f} >= {alpha}) sugiere que este resultado es plausible si la hipótesis nula fuera cierta.")
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
sample_mean = 14  # Media muestral
population_mean_h0 = 15 # Media poblacional bajo H0
sample_std_dev = 1.2 # Desviación estándar muestral (s)
sample_size = 10   # Tamaño de la muestra
alpha = 0.05       # Nivel de significancia
alpha_per_tail = alpha / 2 # Nivel de significancia por cola para prueba bilateral

# Calcular los grados de libertad
degrees_of_freedom = sample_size - 1
print(f"Grados de libertad (df): {degrees_of_freedom}")

# Calcular el error estándar (usando desviación estándar muestral)
# Fórmula: SE = s / sqrt(n)
standard_error = sample_std_dev / np.sqrt(sample_size)
print(f"Error estándar (SE): {standard_error:.4f}")

# Encontrar los valores críticos t para una prueba bilateral con alpha = 0.05 y df
# Valor crítico t de la cola izquierda (área a la izquierda = alpha/2)
critical_t_lower = stats.t.ppf(alpha_per_tail, degrees_of_freedom)
# Valor crítico t de la cola derecha (área a la izquierda = 1 - alpha/2)
critical_t_upper = stats.t.ppf(1 - alpha_per_tail, degrees_of_freedom)

print(f"Valores críticos t para alpha = {alpha} (prueba bilateral) y df = {degrees_of_freedom}: [{critical_t_lower:.4f}, {critical_t_upper:.4f}]")

# Calcular el estadístico de prueba t
# Fórmula: t = (media_muestral - media_poblacional_H0) / SE
t_statistic = (sample_mean - population_mean_h0) / standard_error
print(f"Estadístico t calculado: {t_statistic:.4f}")

# Calcular el valor p para una prueba bilateral
# Para una prueba bilateral, el valor p es el doble del área en la cola
# correspondiente al estadístico de prueba.
# If t_statistic > 0, el valor p = 2 * (1 - stats.t.cdf(t_statistic, degrees_of_freedom))
# If t_statistic < 0, el valor p = 2 * stats.t.cdf(t_statistic, degrees_of_freedom))
p_value = 2 * stats.t.cdf(t_statistic, degrees_of_freedom) # Since t_statistic is likely < 0

print(f"Valor p (prueba bilateral): {p_value:.4f}")

# Graficar la distribución t de Student
x = np.linspace(stats.t.ppf(0.0001, degrees_of_freedom),
                stats.t.ppf(0.9999, degrees_of_freedom), 200)
y = stats.t.pdf(x, degrees_of_freedom)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title(f"Distribución t de Student (df = {degrees_of_freedom}) con Regiones de Rechazo (Bilateral)")
ax.set_xlabel("Valor t")
ax.set_ylabel("Densidad de Probabilidad")
ax.grid(True)

# Rellenar las regiones de rechazo (dos colas)
x_reject_lower = np.linspace(stats.t.ppf(0.0001, degrees_of_freedom), critical_t_lower, 100)
y_reject_lower = stats.t.pdf(x_reject_lower, degrees_of_freedom)
ax.fill_between(x_reject_lower, y_reject_lower, color='purple', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_per_tail})')

x_reject_upper = np.linspace(critical_t_upper, stats.t.ppf(0.9999, degrees_of_freedom), 100)
y_reject_upper = stats.t.pdf(x_reject_upper, degrees_of_freedom)
ax.fill_between(x_reject_upper, y_reject_upper, color='purple', alpha=0.5) # No label for the second fill

# Marcar los valores críticos t
ax.axvline(critical_t_lower, color='purple', linestyle='--', label=f'Valores Críticos t = {critical_t_lower:.4f}, {critical_t_upper:.4f}')
ax.axvline(critical_t_upper, color='purple', linestyle='--')

# Marcar el estadístico t calculado
ax.axvline(t_statistic, color='black', linestyle='-', label=f'Estadístico t Calculado = {t_statistic:.4f}')

# Agregar leyenda
ax.legend()

plt.show()

# Tomar una decisión estadística
print("\n--- Decisión Estadística ---")
if t_statistic < critical_t_lower or t_statistic > critical_t_upper:
    print(f"El estadístico t calculado ({t_statistic:.4f}) cae en la región de rechazo.")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El estadístico t calculado ({t_statistic:.4f}) NO cae en la región de rechazo.")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")

# Tomar una decisión basada en el valor p
print("\n--- Decisión basada en el Valor p ---")
if p_value < alpha:
    print(f"El valor p ({p_value:.4f}) es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El valor p ({p_value:.4f}) NO es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")

# Conclusión en términos prácticos
print("\n--- Conclusión ---")
if p_value < alpha:
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que el consumo promedio real de combustible del auto es significativamente diferente a 15 km/L.")
    print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una media muestral tan extrema o más extrema como 14 km/L (en cualquier dirección), si el verdadero promedio fuera 15 km/L. Un valor p pequeño ({p_value:.4f} < {alpha}) sugiere que este resultado es improbable si la hipótesis nula fuera cierta.")
else:
     print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que el consumo promedio real de combustible del auto es significativamente diferente a 15 km/L. No podemos rechazar la afirmación de que el consumo promedio es de 15 km/L. Esto no significa que la media sea exactamente 15 km/L, sino que la muestra no proporciona evidencia fuerte en contra de H0.")
     print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una media muestral tan extrema o más extrema como 14 km/L (en cualquier dirección), si el verdadero promedio fuera 15 km/L. Un valor p grande ({p_value:.4f} >= {alpha}) sugiere que este resultado es plausible si la hipótesis nula fuera cierta.")
```

Resuelva siguiendo el siguiente protocolo:
1. Plantee las hipótesis nula y alternativa.
2. Calcule el estadístico χ².
3. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de
rechazo.
4. Concluya si se rechaza H0

### 1. Plantear las hipótesis nula y alternativa

La afirmación de la compañía es que la varianza poblacional es de 2.5 minutos². Queremos probar si la varianza poblacional difiere de 2.5.

*   **Hipótesis Nula (H₀):** La varianza poblacional del tiempo de duración de las llamadas es igual a 2.5 minutos². (σ² = 2.5)
*   **Hipótesis Alternativa (H₁):** La varianza poblacional del tiempo de duración de las llamadas es diferente a 2.5 minutos². (σ² ≠ 2.5)

Este es un problema de prueba de hipótesis bilateral (dos colas) para la varianza.

### 2. Establecer el nivel de significancia α

El nivel de significancia dado es α = 0.05.

Para una prueba bilateral, el nivel de significancia se divide entre las dos colas: α/2 = 0.05 / 2 = 0.025.

### 3. Identificar el estadístico de prueba y su distribución

Para pruebas de hipótesis sobre la varianza poblacional, utilizamos el **estadístico Chi-cuadrado (χ²)** y la **distribución Chi-cuadrado**.

### 4. Determinar los grados de libertad, los valores críticos χ² y las regiones de rechazo

Los grados de libertad para la distribución χ² en este caso se calculan como el tamaño de la muestra menos 1: df = n - 1.

Para una prueba bilateral con α = 0.05 y los grados de libertad correspondientes, necesitamos encontrar dos valores χ² críticos: uno en la cola izquierda y uno en la cola derecha.

*   El valor χ² crítico de la cola izquierda es tal que el área a su izquierda es α/2 = 0.025.
*   El valor χ² crítico de la cola derecha es tal que el área a su derecha es 1 - α/2 = 0.975.

La región de rechazo estará en ambas colas de la distribución χ². Rechazaremos H₀ si el estadístico χ² calculado es menor que el valor χ² crítico de la cola izquierda O si el estadístico χ² calculado es mayor que el valor χ² crítico de la cola derecha.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
sample_variance = 4  # Varianza muestral (s²)
population_variance_h0 = 2.5 # Varianza poblacional bajo H0 (σ²)
sample_size = 16   # Tamaño de la muestra (n)
alpha = 0.05       # Nivel de significancia
alpha_per_tail = alpha / 2 # Nivel de significancia por cola para prueba bilateral

# Calcular los grados de libertad
degrees_of_freedom = sample_size - 1
print(f"Grados de libertad (df): {degrees_of_freedom}")

# Calcular el estadístico de prueba Chi-cuadrado
# Fórmula: χ² = ((n - 1) * s²) / σ²
chi_squared_statistic = ((sample_size - 1) * sample_variance) / population_variance_h0
print(f"Estadístico Chi-cuadrado (χ²) calculado: {chi_squared_statistic:.4f}")

# Encontrar los valores críticos Chi-cuadrado para una prueba bilateral con alpha = 0.05 y df
# Valor crítico χ² de la cola izquierda (área a la izquierda = alpha/2)
critical_chi_squared_lower = stats.chi2.ppf(alpha_per_tail, degrees_of_freedom)
# Valor crítico χ² de la cola derecha (área a la izquierda = 1 - alpha/2)
critical_chi_squared_upper = stats.chi2.ppf(1 - alpha_per_tail, degrees_of_freedom)

print(f"Valores críticos χ² para alpha = {alpha} (prueba bilateral) y df = {degrees_of_freedom}: [{critical_chi_squared_lower:.4f}, {critical_chi_squared_upper:.4f}]")

# Calcular el valor p para una prueba bilateral
# Para una prueba bilateral, el valor p es la suma de las áreas en ambas colas
# más allá del estadístico de prueba.
# Si χ² < E[χ²], el valor p = 2 * P(χ² <= χ²_calculado)
# Si χ² > E[χ²], el valor p = 2 * P(χ² >= χ²_calculado)
# E[χ²] = df para la distribución Chi-cuadrado.
expected_chi_squared = degrees_of_freedom

if chi_squared_statistic < expected_chi_squared:
    p_value = 2 * stats.chi2.cdf(chi_squared_statistic, degrees_of_freedom)
else:
    p_value = 2 * (1 - stats.chi2.cdf(chi_squared_statistic, degrees_of_freedom))

print(f"Valor p (prueba bilateral): {p_value:.4f}")


# Graficar la distribución Chi-cuadrado
x = np.linspace(0, stats.chi2.ppf(0.999, degrees_of_freedom) + 5, 200) # Extend x-axis to the right for visualization
y = stats.chi2.pdf(x, degrees_of_freedom)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title(f"Distribución Chi-cuadrado (df = {degrees_of_freedom}) con Regiones de Rechazo (Bilateral)")
ax.set_xlabel("Valor χ²")
ax.set_ylabel("Densidad de Probabilidad")
ax.grid(True)

# Rellenar las regiones de rechazo (dos colas)
x_reject_lower = np.linspace(0, critical_chi_squared_lower, 100)
y_reject_lower = stats.chi2.pdf(x_reject_lower, degrees_of_freedom)
ax.fill_between(x_reject_lower, y_reject_lower, color='purple', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_per_tail})')

x_reject_upper = np.linspace(critical_chi_squared_upper, stats.chi2.ppf(0.999, degrees_of_freedom) + 5, 100)
y_reject_upper = stats.chi2.pdf(x_reject_upper, degrees_of_freedom)
ax.fill_between(x_reject_upper, y_reject_upper, color='purple', alpha=0.5) # No label for the second fill

# Marcar los valores críticos Chi-cuadrado
ax.axvline(critical_chi_squared_lower, color='purple', linestyle='--', label=f'Valores Críticos χ² = {critical_chi_squared_lower:.4f}, {critical_chi_squared_upper:.4f}')
ax.axvline(critical_chi_squared_upper, color='purple', linestyle='--')

# Marcar el estadístico Chi-cuadrado calculado
ax.axvline(chi_squared_statistic, color='black', linestyle='-', label=f'Estadístico χ² Calculado = {chi_squared_statistic:.4f}')

# Agregar leyenda
ax.legend()

plt.show()


# Tomar una decisión estadística
print("\n--- Decisión Estadística ---")
if chi_squared_statistic < critical_chi_squared_lower or chi_squared_statistic > critical_chi_squared_upper:
    print(f"El estadístico χ² calculado ({chi_squared_statistic:.4f}) cae en la región de rechazo.")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El estadístico χ² calculado ({chi_squared_statistic:.4f}) NO cae en la región de rechazo.")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")

# Tomar una decisión basada en el valor p
print("\n--- Decisión basada en el Valor p ---")
if p_value < alpha:
    print(f"El valor p ({p_value:.4f}) es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El valor p ({p_value:.4f}) NO es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")


# Conclusión en términos prácticos
print("\n--- Conclusión ---")
if p_value < alpha:
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente a {population_variance_h0} minutos².")
    print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una varianza muestral tan extrema o más extrema como {sample_variance} minutos² (en cualquier dirección), si la verdadera varianza poblacional fuera {population_variance_h0} minutos². Un valor p pequeño ({p_value:.4f} < {alpha}) sugiere que este resultado es improbable si la hipótesis nula fuera cierta.")
else:
     print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente a {population_variance_h0} minutos². No podemos rechazar la afirmación de que la varianza poblacional es de {population_variance_h0} minutos². Esto no significa que la varianza sea exactamente {population_variance_h0} minutos², sino que la muestra no proporciona evidencia fuerte en contra de H0.")
     print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una varianza muestral tan extrema o más extrema como {sample_variance} minutos² (en cualquier dirección), si la verdadera varianza poblacional fuera {population_variance_h0} minutos². Un valor p grande ({p_value:.4f} >= {alpha}) sugiere que este resultado es plausible si la hipótesis nula fuera cierta.")
```

Resuelva siguiendo el siguiente protocolo:
1. Plantee las hipótesis nula y alternativa.
2. Calcule el estadístico χ²
3. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de

*   **Hipótesis Alternativa (H₁):** La varianza poblacional del llenado es mayor a 1.2 ml². (σ² > 1.2)

Este es un problema de prueba de hipótesis unilateral (cola derecha) para la varianza.

### 2. Establecer el nivel de significancia α

El nivel de significancia dado es α = 0.01.

### 3. Identificar el estadístico de prueba y su distribución

Para pruebas de hipótesis sobre la varianza poblacional, utilizamos el **estadístico Chi-cuadrado (χ²)** y la **distribución Chi-cuadrado**.

La región de rechazo estará en la cola derecha de la distribución χ². Rechazaremos H₀ si el estadístico χ² calculado es mayor que el valor χ² crítico.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
sample_variance = 2.0  # Varianza muestral (s²)
population_variance_h0 = 1.2 # Varianza poblacional bajo H0 (σ²)
sample_size = 20   # Tamaño de la muestra (n)
alpha = 0.01       # Nivel de significancia

# Calcular los grados de libertad
degrees_of_freedom = sample_size - 1
print(f"Grados de libertad (df): {degrees_of_freedom}")

# Calcular el estadístico de prueba Chi-cuadrado
# Fórmula: χ² = ((n - 1) * s²) / σ²
chi_squared_statistic = ((sample_size - 1) * sample_variance) / population_variance_h0
print(f"Estadístico Chi-cuadrado (χ²) calculado: {chi_squared_statistic:.4f}")

# Encontrar el valor crítico Chi-cuadrado para una prueba unilateral de cola derecha con alpha = 0.01 y df
# Usamos stats.chi2.ppf() para obtener el valor χ² cuya área acumulada a la izquierda es 1 - alpha.
critical_chi_squared = stats.chi2.ppf(1 - alpha, degrees_of_freedom)
print(f"Valor crítico χ² para alpha = {alpha} (prueba unilateral cola derecha) y df = {degrees_of_freedom}: {critical_chi_squared:.4f}")

# Calcular el valor p (para una prueba unilateral de cola derecha)
# El valor p es el área a la derecha del estadístico χ² calculado.
p_value = 1 - stats.chi2.cdf(chi_squared_statistic, degrees_of_freedom)
print(f"Valor p: {p_value:.4f}")


# Graficar la distribución Chi-cuadrado
x = np.linspace(0, stats.chi2.ppf(0.999, degrees_of_freedom) + 5, 200) # Extend x-axis to the right for visualization
y = stats.chi2.pdf(x, degrees_of_freedom)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title(f"Distribución Chi-cuadrado (df = {degrees_of_freedom}) con Región de Rechazo (Cola Derecha)")
ax.set_xlabel("Valor χ²")
ax.set_ylabel("Densidad de Probabilidad")
ax.grid(True)

# Rellenar la región de rechazo (cola derecha)
x_reject = np.linspace(critical_chi_squared, stats.chi2.ppf(0.999, degrees_of_freedom) + 5, 100)
y_reject = stats.chi2.pdf(x_reject, degrees_of_freedom)
ax.fill_between(x_reject, y_reject, color='purple', alpha=0.5, label=f'Región de Rechazo (α = {alpha})')

# Marcar el valor crítico Chi-cuadrado
ax.axvline(critical_chi_squared, color='purple', linestyle='--', label=f'Valor Crítico χ² = {critical_chi_squared:.4f}')

# Marcar el estadístico Chi-cuadrado calculado
ax.axvline(chi_squared_statistic, color='black', linestyle='-', label=f'Estadístico χ² Calculado = {chi_squared_statistic:.4f}')

# Agregar leyenda
ax.legend()

plt.show()


# Tomar una decisión estadística
print("\n--- Decisión Estadística ---")
if chi_squared_statistic > critical_chi_squared:
    print(f"El estadístico χ² calculado ({chi_squared_statistic:.4f}) es mayor que el valor crítico χ² ({critical_chi_squared:.4f}).")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El estadístico χ² calculado ({chi_squared_statistic:.4f}) NO es mayor que el valor crítico χ² ({critical_chi_squared:.4f}).")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")

# Tomar una decisión basada en el valor p
print("\n--- Decisión basada en el Valor p ---")
if p_value < alpha:
    print(f"El valor p ({p_value:.4f}) es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, rechazamos la hipótesis nula (H₀).")
else:
    print(f"El valor p ({p_value:.4f}) NO es menor que el nivel de significancia ({alpha}).")
    print("Por lo tanto, no rechazamos la hipótesis nula (H₀).")


# Conclusión en términos prácticos
print("\n--- Conclusión ---")
if p_value < alpha:
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que la varianza poblacional del llenado de botellas es significativamente mayor a {population_variance_h0} ml².")
    print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una varianza muestral tan alta o más alta como {sample_variance} ml², si la verdadera varianza poblacional fuera {population_variance_h0} ml². Un valor p pequeño ({p_value:.4f} < {alpha}) sugiere que este resultado es improbable si la hipótesis nula fuera cierta.")
else:
     print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que la varianza poblacional del llenado de botellas es significativamente mayor a {population_variance_h0} ml². No podemos rechazar la afirmación de que la varianza poblacional es de {population_variance_h0} ml². Esto no significa que la varianza sea exactamente {population_variance_h0} ml², sino que la muestra no proporciona evidencia fuerte en contra de H0.")
     print(f"El valor p ({p_value:.4f}) indica la probabilidad de obtener una varianza muestral tan alta o más alta como {sample_variance} ml², si la verdadera varianza poblacional fuera {population_variance_h0} ml². Un valor p grande ({p_value:.4f} >= {alpha}) sugiere que este resultado es plausible si la hipótesis nula fuera cierta.")
```
