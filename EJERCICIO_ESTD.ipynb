```python
# Valores faltantes
print("\nValores faltantes por columna:")
print(df.isnull().sum())

# Visualización de nulos
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes")
plt.show()
```

NUMÉRICA: Codigo evento, año reporte, edad

CATEGÓRICAS: Nacionalidad, sexo, estrato

```python
df_filtrado_columnas=['CODIGO EVENTO','AÑO REPORTE','EDAD','NACIONALIDAD','SEXO','ESTRATO']
df_filtrado = df[df_filtrado_columnas]

print("\nInformación del dataset filtrado")
print(df_filtrado.info())
print("\nPrimeras filas del dataset filtrado")
print(df_filtrado.head())
```

```python
numerical_cols = ['CODIGO EVENTO', 'AÑO REPORTE', 'EDAD']
df_cleaned = df_filtrado.copy() # Start with the filtered DataFrame

for col in numerical_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out the outliers
    df_cleaned_col = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]

    # Display information about the original and cleaned data for the current column
    print(f"\n--- Análisis de Outliers para '{col}' ---")
    print(f"Shape original de df_filtrado: {df_filtrado.shape}")
    print(f"Q1 ({col}): {Q1}")
    print(f"Q3 ({col}): {Q3}")
    print(f"IQR ({col}): {IQR}")
    print(f"Límite inferior ({col}): {lower_bound}")
    print(f"Límite superior ({col}): {upper_bound}")
    print(f"Número de outliers en '{col}': {df_cleaned.shape[0] - df_cleaned_col.shape[0]}")
    print(f"Shape de df_cleaned_{col}: {df_cleaned_col.shape}")

    # Assign the cleaned DataFrame for the current column to a new variable
    globals()[f'df_cleaned_{col}'] = df_cleaned_col
```

### Histogramas de variables numéricas (sin outliers)

```python
# Histogramas para variables numéricas sin outliers
numerical_cols = ['CODIGO EVENTO', 'AÑO REPORTE', 'EDAD']

colors = {
    'CODIGO EVENTO': '#66BB6A',
    'AÑO REPORTE': '#FB8C00',
    'EDAD': '#E53935'
}

for col in numerical_cols:
    # Access the cleaned DataFrame for the current column
    df_cleaned_col = globals()[f'df_cleaned_{col}']

    plt.figure(figsize=(10, 6))
    sns.histplot(data=df_cleaned_col, x=col, kde=True, color=colors[col])
    plt.title(f'Distribución de {col} (sin outliers)')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

### Visualización para evaluar Normalidad (QQ Plots)

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

numerical_cols = ['CODIGO EVENTO', 'AÑO REPORTE', 'EDAD']

for col in numerical_cols:
    print(f"\n--- QQ Plot para '{col}' ---")
    plt.figure(figsize=(8, 6))
    sm.qqplot(df_filtrado[col], line='s') # 's' - line fit to the data's quantiles
    plt.title(f'QQ Plot de {col}')
    plt.show()
```

### Análisis de Normalidad con diferentes tests (sin gráficos)

```python
from scipy.stats import shapiro, kstest, anderson
import statsmodels.api as sm

numerical_cols = ['CODIGO EVENTO', 'AÑO REPORTE', 'EDAD']

for col in numerical_cols:
    print(f"\n--- Test de Normalidad para '{col}' ---")

    # Shapiro-Wilk Test
    try:
        shapiro_test = shapiro(df_filtrado[col])
        print(f"Shapiro-Wilk Test: Statistic={shapiro_test.statistic:.4f}, p-value={shapiro_test.pvalue:.4f}")
    except Exception as e:
        print(f"Shapiro-Wilk Test could not be performed: {e}")

    # Kolmogorov-Smirnov Test
    try:
        ks_test = kstest(df_filtrado[col], 'norm')
        print(f"Kolmogorov-Smirnov Test: Statistic={ks_test.statistic:.4f}, p-value={ks_test.pvalue:.4f}")
    except Exception as e:
        print(f"Kolmogorov-Smirnov Test could not be performed: {e}")

    # Anderson-Darling Test
    try:
        anderson_test = anderson(df_filtrado[col])
        print(f"Anderson-Darling Test: Statistic={anderson_test.statistic:.4f}")
        for i in range(len(anderson_test.critical_values)):
            sl, cv = anderson_test.significance_level[i], anderson_test.critical_values[i]
            print(f"  Significance Level {sl}%: Critical Value={cv:.4f}")
    except Exception as e:
        print(f"Anderson-Darling Test could not be performed: {e}")

    # Jarque-Bera Test
    try:
        jb_test = sm.stats.stattools.jarque_bera(df_filtrado[col])
        print(f"Jarque-Bera Test: Statistic={jb_test[0]:.4f}, p-value={jb_test[1]:.4f}, Skewness={jb_test[2]:.4f}, Kurtosis={jb_test[3]:.4f}")
    except Exception as e:
        print(f"Jarque-Bera Test could not be performed: {e}")
```
